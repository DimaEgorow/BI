# Поиск зависимостей

# 1. Построив дашборд и увидев графики я выявил несколько устойчивых закономерностей(и описал их)

1.1 Первая зависимость это зависимость количества вакансий от региона, во все месяца устойчивые показатели проявляет город Москва с наибольших количеством вакансий, а наименьшие показатели показывает Краснодарский край.\
1.2 Вторая зависимость это зависимость зароботной платы от опыта работы, работники с опытом более 6 лет могут расчитывать на более высокую зарплату, а работники без опыта работы на самую минимальную по рынку\
1.3 Третья зависимость эта зависимость заработной платы от региона, на самую большую зарплату можно расчитывать в городе Москва, а на самую мимнимальную в Краснодарском крае и Свердловской области
