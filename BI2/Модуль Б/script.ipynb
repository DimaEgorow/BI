# Выполнение анализа зависимости между доходом и уровнем образования(Опытом работы)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла XLSX
data = pd.read_excel('finishdata.xlsx')

# Проведение анализа зависимости между доходом и уровнем образования
income_education_analysis = data.groupby('Опыт работы')['Заработная плата'].mean()

# Построение графика зависимости
plt.figure(figsize=(10,6))
income_education_analysis.plot(kind='bar', color='skyblue')
plt.title('Зависимость дохода от уровня образования(Опыта работы)')
plt.xlabel('Уровень образования(Опыт работы)')
plt.ylabel('Средний доход')
plt.xticks(rotation=45)
plt.show()
```
